<h3>Calculating the retention rate</h3>
<h4>Retention rate by beginning and end date</h4>
<p>
Retention Rate = <br>
(Number of customers at the end of the period / Number of customers at the beginning of the period) * 100
</p>

```python
# marketing.info()
```

```python
print("Beginning / end of the period")
print('------------------------------------------------------------------------------')
print(marketing['date_served'].min())
print(marketing['date_served'].max())
```

```python
market_min = \
marketing[(marketing['date_served']==marketing['date_served'].min()) 
         ] [['date_served', 'is_retained']]

market_max = \
marketing[(marketing['date_served']==marketing['date_served'].max()) 
         ] [['date_served', 'is_retained']]

market_min_max = pd.concat([market_min, market_max])



print("True: retained, False: not retained")
print('------------------------------------------------------------------------------')
print(market_min_max.value_counts('is_retained'))

print(683+39)
print(375+347)
```

<h4>
Number of customers at the end of the period / <br> Number of customers at the beginning of the period
</h4>

```python
retained_beginning_end_date = \
market_min_max.groupby('date_served')['is_retained'].count()

print(retained_beginning_end_date)
```

```python
retention_rate = \
round(retained_beginning_end_date.values[1] / 
      retained_beginning_end_date.values[0] * 100, 2)

print(f'retention rate: {retention_rate} %')
```

<h4>Retention rate by subscribers divided by total retained for every channel</h4>
<h5>Group and count by multiple cols based on conditions</h5>

```python
# Count the subs by subscribing channel and day
# denominator
retention_total = marketing.groupby(['date_served',
                                     'subscribing_channel'])['user_id'].nunique()

# Print results
print(retention_total.head())
```

```python
# Sum the retained subs by subscribing channel and date subscribed
# nominator
retention_subs = \
marketing[marketing['is_retained']==True].\
groupby(['date_served', 'subscribing_channel'])['user_id'].nunique()

# Print results
print(retention_subs.head())
```

```python
# Divide retained subscribers by total subscribers
retention_rate = retention_subs/retention_total
print(retention_rate.tail())
```

<p>

```python
sum_of_second_index = retention_rate.index.levels[1].tolist()
print(sum_of_second_index)
```

```python
email_sum = retention_rate.loc[slice(None), 'Email']

print(email_sum.values.tolist()[:3])
print(email_sum.values.sum())
# sum() is calculating the total sum of all values
print(email_sum.values[0])
# Which is the first value in the Series.
```

```python
retention_rate_df = pd.DataFrame(retention_rate.unstack(level=1))
print(retention_rate_df.tail())
```

<h4>
Row- and col sums
</h4>

```python
retention_rate_df_sums = retention_rate_df.copy()
print(type(retention_rate_df_sums))
retention_rate_df_sums['row_sum'] = retention_rate_df_sums.sum(axis=1)
# df.loc['Total'] = df.sum()
retention_rate_df_sums.loc['Total'] = retention_rate_df_sums.sum().round(2)
```
